```python
from bs4 import BeautifulSoup 
import requests
import csv
import pandas as pd
```

```python
url ="https://www.flipkart.com/search?q=oneplus+mobile"
```

```python
req=requests.get(url)
```

```python
req
```

```python
soup=BeautifulSoup(req.content,'html.parser')
```

```python
soup
```

```python
titles=soup.find_all('div',class_='_4rR01T')
print(titles)
```

```python
ratings=soup.find_all('div',class_='_3LWZlK')
print(ratings)
```

```python
prices=soup.find_all('div',class_='_30jeq3 _1_WHN1')
```

```python
print(prices)
```

```python
reviews=soup.find_all('span',class_='_2_R_DZ')
```

```python
print(reviews)
```

```python
mt=[]
mr=[]
mp=[]
mre=[]
```

```python
for title,rating,price,rev in zip(titles,ratings,prices,reviews):
    mt.append(title.text)
    mr.append(rating.text)
    mp.append(price.text)
    mre.append(rev.text)
```

```python
print(mt)
```

```python
print(mr)
```

```python
print(mp)
```

```python
print(mre)
```

saving data in csv format


```python
d={'mt':mt,'mr':mr,'mp':mp,'mre':mre}
```

```python
model=pd.DataFrame(data=d)
```

```python
model.to_csv("mobilesdata.csv")
```
